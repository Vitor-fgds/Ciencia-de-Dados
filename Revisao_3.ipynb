05) Porquê os comandos df.iloc[0:2] e df.loc[0:2] retornam resultados distintos?

```python
df.iloc[0:2]
df.loc[0:2]
```

Resposta 05) O comando "iloc" exclui o último índice, ou seja, em iloc[0:2] os elementos de índice 2 não serão mostrados. Enquanto isso, o comando "loc" mostra todos os índices, ou seja, no comando loc[0:2] os elementos de índice 0,1,2 serão mostrados
