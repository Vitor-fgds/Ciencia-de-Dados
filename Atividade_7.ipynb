```python
import pandas as pd
df = pd.read_csv('/content/precos-semestrais-ca-2022-01.csv', delimiter=';')
import matplotlib.pyplot as plt
df

```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.')
```

```python
df['Valor de Venda'] = df['Valor de Venda'].fillna(0)
```

```python
df['Valor de Venda'] = df['Valor de Venda'].astype('float64')
```

```python
x = df[df['Produto'] == 'GASOLINA']
y = df[df['Produto'] == 'ETANOL']

media_estado = x.groupby('Estado - Sigla')['Valor de Venda'].mean()


```

```python
plt.figure(figsize=(15,5))
plt.plot(media_estado, marker='o')
plt.show()
```

```python
media_estado = x.groupby('Estado - Sigla')['Valor de Venda'].mean()
maiores_medias = media_estado.sort_values().tail(10)
plt.figure(figsize=(15,5))
plt.plot(maiores_medias, marker='o')
plt.show()
```

```python
media_estado = y.groupby('Estado - Sigla')['Valor de Venda'].mean()
maiores_medias = media_estado.sort_values().head(10)
plt.figure(figsize=(15,5))
plt.plot(maiores_medias, marker='o')
plt.show()
```
