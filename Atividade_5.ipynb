```python
2) Conte quantos jogos cada pais fez em copas do mundo.
```

```python
jogos_casa = jogos['HomeTeam'].value_counts()
jogos_visitante = jogos['AwayTeam'].value_counts()
jogos_total = jogos_casa + jogos_visitante
print(jogos_total)
```

4) Conte o total de gols que foram feitos em copas;

```python
gols_casa = jogos['HomeGoals'].sum()
gols_fora = jogos['AwayGoals'].sum()
gols_total = gols_casa + gols_fora
print(gols_casa)
print(gols_fora)
print(f' O numero total de gols foi: {gols_total}')
```

5) Apresente o país que fez mais gols em copas; (groupby)

```python
gols_casa = jogos[['HomeTeam', 'HomeGoals']].groupby('HomeTeam').sum()
gols_fora = jogos[['AwayTeam', 'AwayGoals']].groupby('AwayTeam').sum()
gols_total = gols_fora['AwayGoals'] + gols_casa['HomeGoals']
gols_total.fillna(0, inplace=True)
print(gols_total.sort_values().tail(1))
```

```python
6) Qual foi a maior goleada? abs() retorna apenas valores positivos
```

```python
jogos['saldo_de_gols'] = (jogos['HomeGoals'] - jogos['AwayGoals']).abs()
jogos.loc[jogos['saldo_de_gols'] == jogos['saldo_de_gols'].max()]
```
